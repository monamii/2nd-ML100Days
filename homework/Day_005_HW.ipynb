```python
incomeDf = app_train['AMT_INCOME_TOTAL']
```

```python
incomeDf.describe()
```

```python
print('Variance: %d' % incomeDf.var())
print('sqrt(Variance): %d' % math.sqrt(incomeDf.var()))
print('Standard deviation: %d' % incomeDf.std())
```

```python
print('Skewness: %d' % incomeDf.skew())
print('Kurtosis: %d' % incomeDf.kurt())
```
